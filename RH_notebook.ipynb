```python
import pandas as pd
```

```python
df1 = pd.read_csv('By_state.csv')
```

```python
df1.head()
```

```python
df1.Total_population = df1.Total_population.str.replace(',','')
```

```python
df1.RH_population = df1.RH_population.str.replace(',', '')
```

```python
df1.Total_population = df1.Total_population.astype(int)
```

```python
df1.RH_population = df1.RH_population.astype(int)
```

```python
df1['RH_c_percentage'] = (df1.RH_population/df1.Total_population)*100
```

```python
df1.sort_values('RH_population', ascending = True)
```

```python
df_top_5 = df1.sort_values('RH_c_percentage', ascending = False).head(5)
df_bot_5 = df1.sort_values('RH_c_percentage', ascending = True).head(5)
```

```python
ax = df_top_5.plot.barh(x='Jurisdiction', y="RH_c_percentage")
```

```python
ax = df_bot_5.plot.barh(x='Jurisdiction', y="RH_c_percentage")
```
